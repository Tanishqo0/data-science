```python
x = [1,2,3,4,5]
y = [1,10,9,30,25]
plt.plot(x,y,marker = "D", ms = 10, mfc = "red", mec = "green",
        color = "purple", lw = 3, ls = ":",
        alpha = 0.3, label = "lineplot")    # ms=Marker size  , mfc= marker face color  , 
                                            #   mec = marker edge color , lw= line width,  ls = line style
plt.title("lineplot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours, hatch = hatch_patterns, label = x)
plt.legend()
plt.show()
```

```python
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
colours = ["red","green","blue","yellow"]
hatch_patterns = ["//","\\\\","xx","oo"]
plt.bar(x,y,color = colours, hatch = hatch_patterns,
        label = x, width=0.3, edgecolor = "black", alpha = 0.7,
        linewidth = 3)
plt.legend()
plt.show()
```

```python
## histrogram -- distribution of data in interval(bins)
arr = np.random.randiat(20000,40000,100)
plt.hist(arr,bins=20)
plt.show()
```

```python
## piechart -- percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x)
plt.show()
```

```python
## piechart -- percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
plt.pie(y,labels = x, autopct = "%1.2f%%", shadow =True)
plt.show()
```

```python
## piechart -- percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = (0.2,0,0,0)
plt.pie(y,labels = x, autopct = "%1.2f%%", shadow =True, explode = explodes)
plt.show()
```

```python
# piechart -- percentage
x = ["mp","up","rj","gj"]
y = [10,20,30,40]
explodes = (0.2,0,0,0)
plt.pie(y,labels = x, autopct = "%1.2f%%", shadow =True, explode = explodes)
plt.savefig("piechart.pdf")
plt.show()
```

```python
##subplot
x = np.linspace(0,10,100)
plt.figure(figsize=(10,6))
plt.subplot(2,2,1) 

plt.plot(x,np.sin(x))
plt.subplot(2,2,2) 
plt.plot(x,np.cos(x))
plt.subplot(2,2,3) 
plt.plot(x,np.tan(x))
plt.subplot(2,2,4) 
plt.plot(x,np.log(x))
plt.tight_layout()
plt.show()
```
